```python
import cv2
import numpy as np
import os
import pandas as pd
```

```python
image_folder = r"C:\Users\Shreyas R . P\Desktop\acm\block data\block 3"
output_csv = "datablockthree.csv"
```

```python
image_files = list()
for f in os.listdir(image_folder):
    if f.endswith((".jpg",".png")):
        image_files.append(f)
data = []
```

```python
for image_file in image_files:
    image_path = os.path.join(image_folder,image_file)
    image = cv2.imread(image_path)
    if image is None:
        print(f" Error loading {image_file}. Skipping...")
        continue
    rgb_value = cv2.cvtColor(image , cv2.COLOR_BGR2RGB)
    pixels = rgb_value.reshape(-1,3)
    filtered_values=[]
    for v in pixels:
        if not np.all(v == 0):
            filtered_values.append(v.tolist())
    data.append([image_file, filtered_values])
```

```python
df = pd.DataFrame(data, columns=["Image", "filtered_values"])
df.to_csv(output_csv, index=False)
```
